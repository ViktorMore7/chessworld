```python
import camelot
import os
import shutil
from google.colab import drive

# --- SETTINGS ---
pdf_path = '/content/drive/MyDrive/Colab Notebooks/Colab_table/_pefo-2019-appendix-a.pdf'
output_folder = '/content/drive/MyDrive/Colab Notebooks/Colab_table/extracted_tables'

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

print(f"Analyzing PDF for tables: {os.path.basename(pdf_path)}...")

# 1. Extract tables
# 'flavor' can be 'lattice' (for tables with lines) or 'stream' (for tables without lines)
tables = camelot.read_pdf(pdf_path, pages='all', flavor='lattice')

print(f"Total tables found: {len(tables)}")

# 2. Save each table as a separate CSV file
for i, table in enumerate(tables):
    # Table indexing starts at 0, but we'll save as Table_1, Table_2, etc.
    table_filename = f"table_{i+1}.csv"
    save_path = os.path.join(output_folder, table_filename)

    # Export to CSV
    table.to_csv(save_path)

    # Preview the first few rows in the console
    print(f"âœ… Saved Table {i+1} to {table_filename}")
    print(table.df.head(3)) # Show top 3 rows of each found table
    print("-" * 30)

print(f"\nAll tables are now in: {output_folder}")
```

```python
import camelot
import os

# Paths for your specific files
folder_path = '/content/drive/MyDrive/Colab Notebooks/Colab_table'
files_to_process = [
    '_ecb.pr200608_ssi_table~3054d55051.en.pdf',
    '_pefo-2019-appendix-a.pdf'
]

output_folder = os.path.join(folder_path, 'extracted_tables')
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for file_name in files_to_process:
    full_path = os.path.join(folder_path, file_name)
    print(f"--- Processing: {file_name} ---")

    # WE SWITCHED TO flavor='stream' BECAUSE THESE TABLES LACK BORDERS
    # pages='all' ensures we don't miss any pages
    tables = camelot.read_pdf(full_path, pages='all', flavor='stream')

    print(f"Found {len(tables)} tables.")

    for i, table in enumerate(tables):
        clean_name = file_name.replace('.pdf', '')
        save_path = os.path.join(output_folder, f"{clean_name}_table_{i+1}.csv")
        table.to_csv(save_path)
        print(f"  âœ… Saved Table {i+1}")

print("\nProcessing complete. Check your 'extracted_tables' folder.")
```

```python
import camelot
import os

# --- SETTINGS ---
# The folder where your uploaded files are located
folder_path = '/content/drive/MyDrive/Colab Notebooks/Colab_table'
output_folder = os.path.join(folder_path, 'extracted_tables')

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# List of your specific files
my_files = [
    '_ecb.pr200608_ssi_table~3054d55051.en.pdf',
    '_pefo-2019-appendix-a.pdf'
]

for file_name in my_files:
    full_path = os.path.join(folder_path, file_name)
    print(f"\nðŸ” Analyzing: {file_name}")

    # We use 'stream' because these tables have no borders
    # We use 'pages="all"' to ensure we catch every financial statement
    tables = camelot.read_pdf(full_path, pages='all', flavor='stream')

    print(f"ðŸ“ˆ Found {len(tables)} tables.")

    for i, table in enumerate(tables):
        # Create a unique name for each table found
        base_name = os.path.splitext(file_name)[0]
        csv_name = f"{base_name}_table_{i+1}.csv"
        save_path = os.path.join(output_folder, csv_name)

        # Save to CSV
        table.to_csv(save_path)
        print(f"   âœ… Saved Table {i+1} to {csv_name}")

print("\n--- Process Finished ---")
```
